```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')

from Recordings import Recordings
import mne
from Recordings import AEPFeedbackRecording, Recordings, Recording
import numpy as np
import os
```

```python
# Load the data
basepath = './Recordings'
recordings = Recordings(basepath)
aep_feedback_recordings = recordings.filter_by(experiment_id='aep_feedback')
aep_recordings = recordings.filter_by(experiment_id='aep')
```

```python
def plot_subjects(recordings, subfolder=None, name_pattern='output'):
    for i in range(1, 8):
        subject_recordings = recordings.filter_by(subject_id=i)
        # Merge epochs of all recordings of the same subject
        epochs = []
        for recording in subject_recordings.recordings:
            epochs.append(recording._epochs)
        epochs = mne.concatenate_epochs(epochs)
        evokeds = dict(
            standard=list(epochs["standard"].iter_evoked()),
            oddball=list(epochs["oddball"].iter_evoked()),
        )
        fig = mne.viz.plot_compare_evokeds(evokeds, combine="mean", title=f'Subject {i}')[0]
        if subfolder is None:
            fig.savefig(f'./images/{name_pattern}_{i}.png')
        else:
            os.makedirs(os.path.join('./images', subfolder), exist_ok=True)
            fig.savefig(os.path.join('./images', subfolder, f'{name_pattern}_{i}.png'))
```

```python
plot_subjects(aep_recordings, 'aep', '')
```
